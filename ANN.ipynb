```python
import cv2
import os
```

```python
import numpy as np
import mediapipe as mp
mp_pose = mp.solutions.pose
pose = mp_pose.Pose()
mp_drawing = mp.solutions.drawing_utils
```

```python
def check_image(image_path):
    if os.path.exists(image_path):
        image = cv2.imread(image_path)
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        results = pose.process(image_rgb)

        if results.pose_landmarks:
            return 1
        else:
            print(f"Cannot read: {image_path}")
            return 403
    else:
        print(f"Missing: {image_path}")
        return 404
```

```python
def extract_pose_landmarks(image_path):
    # Read the image frame
    if os.path.exists(image_path):
        image = cv2.imread(image_path)

        # Convert the image to RGB
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Process the image and extract hand landmarks
        results = pose.process(image_rgb)

        if results.pose_landmarks:  # !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!  need add zeros in else condition
            pose_landmark = results.pose_landmarks  # Get the first detected hand
            keypoints = []
            for landmark in pose_landmark.landmark:
                keypoints.append([landmark.x, landmark.y, landmark.z])

            return np.array(keypoints).flatten()
        else:
            print(f"Cannot read: {image_path}")
            return 403
    else:
        print(f"Missing: {image_path}")
        return 404
```

```python
import glob
yoga_poses = ['downdog','goddess','plank','tree','warrior']

image_path = 'C:\\Users\\AR Ankon\\Desktop\\Ankon3\\DATASET\\'
max_no_images = 1000
max_no_data = 99999
supported_extensions = ['*.jpg', '*.jpeg', '*.png', '*.bmp']  # Add any other image formats as needed

for yoga_pose in yoga_poses:
    count = 0
    for ext in supported_extensions:
        image_files = glob.glob(os.path.join(image_path, yoga_pose, ext))  # Get all files matching the extension
        for impath in image_files[:max_no_images]:  # Limit to max_no_images
            value = check_image(impath)
            if value == 1:
                count += 1

    if count < max_no_data:
        max_no_data = count

print(f"Max number of data: {str(max_no_data)}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import TensorBoard, ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.models import load_model
```

```python
log_dir = os.path.join('Logs')
tb_callback = TensorBoard(log_dir=log_dir)
```

```python
model = Sequential()
model.add(Dense(128, activation='tanh', input_shape=(99,)))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='tanh'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='tanh'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='tanh'))
model.add(Dense(np.array(yoga_poses).shape[0], activation='softmax'))
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
# checkpoint = ModelCheckpoint(filepath='C:/Users/anikm/Desktop/Thesis/', monitor='val_acc', save_best_only=True, mode='max')
earlyStopping = EarlyStopping(monitor='categorical_accuracy', patience=10, verbose=0, mode='min')
filepath = 'model_weights.keras'
mcp_save = ModelCheckpoint(filepath, save_best_only=True, monitor='categorical_accuracy', mode='max') 

reduce_lr_loss = ReduceLROnPlateau(monitor='categorical_accuracy', factor=0.1, patience=7, verbose=1, min_delta=1e-4, mode='min')
```

```python
model.fit(X_train, y_train, epochs=500, callbacks=[mcp_save, reduce_lr_loss, tb_callback])
```
